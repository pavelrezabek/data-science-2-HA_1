```python
#imports
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt


from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix

from sklearn import tree
from sklearn.ensemble import RandomForestRegressor

import sklearn

import CHAID


 
```

```python
train_data = pd.read_csv("2023_DS2_HW1_data_train.csv")
```

```python
train_data
```

```python
data_cart = train_data.copy()
data_cart.isna().sum()
```

```python
data_cart.dtypes
```

```python
data_cart.fillna(value= -1, inplace=True)
data_cart = data_cart[data_cart['booking_status'] != -1]

```

```python
#TBD: pridat 'type_of_meal_plan','room_type_reserved',market_segment_type',

target = 'booking_status'

predictors = [ 'no_of_adults', 'no_of_children', 'no_of_weekend_nights',
       'no_of_week_nights',  'required_car_parking_space',
        'lead_time', 'arrival_year', 'arrival_month',
       'arrival_date', 'repeated_guest',
       'no_of_previous_cancellations', 'no_of_previous_bookings_not_canceled',
       'avg_price_per_room', 'no_of_special_requests']
```

```python
data_cart_train, data_cart_test = train_test_split(
    data_cart, test_size=0.2, random_state=17
)
```

```python
criterion = 'gini' #(TODO)
splitter = 'best' #(TODO)
mdepth = 20 #(TODO)
min_samples_split = 5 #(TODO)
min_samples_leaf = 0.001 #(TODO)
minleaf = 100 #(TODO)
random_state = 17


model = tree.DecisionTreeClassifier(
    criterion=criterion,
    splitter=splitter, 
    max_depth=mdepth,
    min_samples_split=min_samples_split,
    min_samples_leaf=min_samples_leaf, 
    random_state=random_state, 
)
```

```python

# And fit training data.

clf = model.fit(data_cart_train[predictors], data_cart_train['booking_status'])

```

```python
xxx
```

```python
pred_labels_tr = model.predict(data_cart_train[predictors])
pred_labels_te = model.predict(data_cart_test[predictors])
```

```python
# Tree summary and model evaluation metrics
print('*************** Tree Summary ***************')
print('Classes: ', clf.classes_)
print('Tree Depth: ', clf.tree_.max_depth)
print('No. of leaves: ', clf.tree_.n_leaves)
print('No. of features: ', clf.n_features_)
print('--------------------------------------------------------')
print("")

print('*************** Evaluation on Test Data ***************')
score_te = model.score(data_cart_test[predictors], data_cart_test[target])
print('Accuracy Score: ', score_te)
# Look at classification report to evaluate the model
print(classification_report(data_cart_test[target], pred_labels_te))
print('--------------------------------------------------------')
print("")

print('*************** Evaluation on Training Data ***************')
score_tr = model.score(data_cart_train[predictors], data_cart_train[target])
print('Accuracy Score: ', score_tr)
# Look at classification report to evaluate the model
print(classification_report(data_cart_train[target], pred_labels_tr))
print('--------------------------------------------------------')
```

```python
fig, ax = plt.subplots(figsize=(25, 25))
tree.plot_tree(clf, ax=ax)
plt.show()
```

```python
data_cart_test['Prediction_prob'] = clf.predict_proba(data_cart_test[predictors])[:,1] #(TODO)
data_cart_test['Prediction'] = (data_cart_test['Prediction_prob']>=0.5).astype(int) #(TODO)




data_cart_train['Prediction_prob'] = clf.predict_proba(data_cart_train[predictors])[:,1] #(TODO)
data_cart_train['Prediction'] = (data_cart_train['Prediction_prob']>=0.5).astype(int) #(TODO)
```

```python
conf_mx = pd.DataFrame(sklearn.metrics.confusion_matrix(data_cart_test["booking_status"], data_cart_test['Prediction']))
conf_mx
#v řádku predikce, v sloupci realita
```

```python
# Get numerical feature importances
importances = list(clf.feature_importances_)
# List of tuples with variable and importance
feature_importances = [(feature, round(importance, 2)) for feature, importance in zip(predictors, importances)]
# Sort the feature importances by most important first
feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)
# Print out the feature and importances 
[print('Variable: {:20} Importance: {}'.format(*pair)) for pair in feature_importances];
```

```python
# Plot importance
x_values = list(range(len(importances)))
# Make a bar chart
plt.bar(x_values, importances, orientation = 'vertical')
# Tick labels for x axis
plt.xticks(x_values, predictors, rotation='vertical')
# Axis labels and title
plt.ylabel('Importance'); plt.xlabel('Variable'); plt.title('Variable Importances');
```

```python
fpr, tpr, thresholds = roc_curve(data_cart_test["booking_status"], data_cart_test['Prediction_prob'])
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()


auc = roc_auc_score(data_cart_test["booking_status"], data_cart_test['Prediction_prob'])
print("AUC:", auc)
```
